## Imports

```python
import kagglehub
import pandas as pd
import requests
```

## 1. AWS SaaS Sales Dataset (Kaggle) 

```python
# Download Dataset from Kaggle
path = kagglehub.dataset_download("nnthanh101/aws-saas-sales")

# Load into DataFrame
df_saas = pd.read_csv(f"{path}/SaaS-Sales.csv")

# Save raw data locally
df_saas.to_csv("../data/raw/saas_sales.csv", index=False)

df_saas.shape
```

## 2. World Bank - GDP per Capita
**Explanation:** World Bank API provides economical data for every country, no fees, no API key. 
We need 2 indicators (indicators):

  - NY.GDP.PCAP.CD = GDP per Capita in US-Dollar                                                                                                                                                                                                                      
  - IT.NET.USER.ZS = Internet-Users in % of the population 

```python
# Fetch GDP per capita (current US$) for all countries, most recent year
url_gdp = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.CD?format=json&per_page=300&date=2020:2023&mrnev=1" 
response_gdp = requests.get(url_gdp)
data_gdp = response_gdp.json()[1]       # API provides Array with 2 elements: [0] = metadat, [1] = actual data

# Parse into DataFrame
df_gdp = pd.DataFrame([
    {"Country": item['country']['value'], "GDP_per_Capita": item['value']}
    for item in data_gdp if item['value'] is not None       # filter for countries with no data
])

df_gdp.to_csv("../data/raw/gdp_per_capita.csv", index=False)
df_gdp.shape
```

## 3. World Bank - Internet Penetration

```python
# Fetch internet users (% of population) for all countries, most recent year
url_net = "https://api.worldbank.org/v2/country/all/indicator/IT.NET.USER.ZS?format=json&per_page=300&date=2020:2023&mrnev=1"                                                                                                                                       
response_net = requests.get(url_net)                                                                                                                                                                                                                                
data_net = response_net.json()[1]

# Parse into DataFrame
df_net = pd.DataFrame([
    {"Country": item['country']['value'], "Internet_Penetration": item['value']}
    for item in data_net if item['value'] is not None
])

df_net.to_csv("../data/raw/internet_penetration.csv", index=False)
df_net.shape

# NOTE: same pattern, just different indicator; same procedure every time, get as much indicators as needed
```
